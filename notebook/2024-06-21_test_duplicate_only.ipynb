```python
%cd /kaggle/working
import pandas as pd
import polars as pl
```

```python
test_df = pl.read_csv("input/leap-atmospheric-physics-ai-climsim/test.csv")
test_old_df = pl.read_csv("input/leap-atmospheric-physics-ai-climsim/test_old.csv")
```

```python
submission_df = pl.read_parquet(
    "gs://kaggle-leap/kami/experiments/204_diff_last/all_lr/submission.parquet"
)
```

```python
submission_df.head()
```

```python
tmp = test_df[:, 1::5].with_row_index("index")
tmp_old = test_old_df[:, 1::5].with_row_index("old_index")
merge_df = pd.merge(tmp.to_pandas().round(6), tmp_old.to_pandas().round(6))
```

```python
merge_df
```

```python
duplicated_index = merge_df["index"]
duplicated_index[:10]
```

```python
# 選択以外の
submission_df = (
    submission_df.with_row_index("index")
    .with_columns(
        [
            pl.when(~pl.col("index").is_in(duplicated_index))
            .then(0)
            .otherwise(pl.col(col_name))
            .alias(col_name)
            for col_name in submission_df.columns[1:]
        ]
    )
    .drop(["index"])
)
```

```python
submission_df.write_parquet(
    "output/204_diff_last_all_lr_submission_only_duplicate.parquet"
)
```
