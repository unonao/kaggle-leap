```python
%cd /kaggle/working
import polars as pl

df = pl.read_parquet("input/train.parquet", n_rows=30000)
```

```python
first_location = df[::384]
first_location = (
    first_location.select(["pbuf_SOLIN"])
    .with_row_index("index")
    .with_columns(
        [
            ((pl.col("index") * 1200 * 7) % (3600 * 24)).alias("time_seconds"),
            ((pl.col("index") * 1200 * 7) // (3600 * 24)).alias("date"),
        ]
    )
)

first_location
```

```python
with pl.Config(tbl_rows=60):
    print(first_location.head(60))
```

```python
test_df = pl.read_parquet("input/test.parquet", n_rows=20000)
```
